**DATA COLLECTION**

The dataset was gotten from Draky Supermarket in the form of a csv file and it contains 2994 rows and 14 columns.
These are the column names and brief description of each:

1.  Product_Identifier: The product identifier for each product

2.  Supermarket Identifier: The supermarket identifier for each product

3.  Product_Supermarket Identifier: The unique identifier for each entry in the dataset

4.  Product_Weight: The weight of the product

5.  Product_Fat_Content:  Indicates the level or amount of fat content in the product

6.  Product_Shelf_Visibility: Represents the visibility of the product on the shelf in the           supermarket

7.  Product_Type: Categorizes the product into different types or categories

8.  Product_Price: The price of the product

9.  Supermarket_Opening_Year: The year in which the supermarket was opened

10. Supermarket_Size: Represents the size or capacity of the supermarket

11. Supermarket_Type: Indicates the type of location where the supermarket is situated

12. Supermarket_Location_Type: Categorizes the supermarket based on its type

13. Average_Price_Per_Product_Type: Average price of each product within the same product type

14. Product_Supermarket_Sales: The sales amount of the product

```python
# to approximate and round up the decimal values

# the apply()function is used to apply the specified lambda function to each value in the column.
# The lambda funtion approximates the value using the 'decimal' module, quantizes it to two decimal
# places with rounding to the nearest even value, and then applies the round() function to round
# the approximated value to two decimal places.

df['Product_Weight_Kg'] = df['Product_Weight_Kg'].apply(lambda x: round(decimal.Decimal(str(x)).quantize(decimal.Decimal('0.00'), rounding=decimal.ROUND_HALF_UP), 2))
```

**Product Price vs Product Sales**

I utilized a line chart to examine the relationship between Product Price and product sales and determine whether there is an impact of price on sales.

```python
# This line uses the NumPy polyfit function to fit a linear regression line to the data points.
# It takes two arrays as inputs: df['Product_Price'] as the x-values and df['Product_Supermarket_Sales'] as the y-values.
# The number 1 represents the degree of the polynomial fit, which in this case is a first-degree polynomial (a straight line).
# The function returns the slope m and the y-intercept b of the regression line.

m, b = np.polyfit(df['Product_Price'], df['Product_Supermarket_Sales'], 1);

# This line plots the original data points as scatter markers ('o') on the chart.
# It uses the df['Product_Price'] as the x-values and df['Product_Supermarket_Sales'] as the y-values.

plt.plot(df['Product_Price'], df['Product_Supermarket_Sales'], 'o');

# This line plots the regression line using the equation y = mx + b, where m is the slope and b is the y-intercept.
# It uses the df['Product_Price'] as the x-values and m * df['Product_Price'] + b as the y-values.

plt.plot(df['Product_Price'], m * df['Product_Price'] + b);

# set the labels

plt.title('Product Prices vs. Sales');
plt.xlabel('Product Price');
plt.ylabel('Product Sales');
```

we can see a positive slope which indicates a positive relationship between the variables. We can assume that as the 'Product_Price' increases, the 'Product_Supermarket_Sales' also tends to increase.
However since the slope is almost flat, I decided to plot a regression analysis so as to draw more accurate insights from the variables.

```python
# import the module

import statsmodels.api as sm
```

```python
# I created a dataframe with the variables

df_regression = df[['Product_Price', 'Product_Supermarket_Sales']]
```

```python
# to view the dataframe

df_regression.sample(10)
```

```python
# add a constant column with a value of 1 to the DataFrame df_regression.
# By including a constant term, the regression model can estimate both the slope (coefficients)
# and the intercept (constant term) when fitting the data.

df_regression = df_regression.assign(constant=1)
```

```python
# assign product supermarket sales to the dependant variable 'y'
y = df['Product_Supermarket_Sales']

# assign product price to the independent variable 'x'
# The .copy() method is used to make a copy of the column, ensuring that any changes made to X do
# not affect the original DataFrame.

X = df_regression[['Product_Price']].copy()
X['constant'] = 1
```

```python
# OLS means ordinary least squares
model = sm.OLS(y, X)
results = model.fit()
```

```python
print(results.summary())
```

These results suggest that the product price has a positive influence on sales and the higher the sales price, the higher the sales but it does not fully explain the variation observed and there may be other factors beyond the product price may contribute to sales.
